#### 4. 당시에는 월~일을 모두 예측 데이터로 활용했기 때문에 토/일요일의 예측값이 터무니없게 나오는 문제가 있었습니다. 이를 별도의 조정 상수를 곱해 처리했습니다.

```python
forecastProphetTable['weekday']= forecastProphetTable['ds'].apply(lambda x: x.weekday())
```

```python
forecastProphetTable['weekday'].replace([0,1,2,3,4,5,6], [1, 1, 1, 1, 1, 0.734814572, 0.005993335], inplace= True)
```

```python
forecastProphetTable['yhat']= forecastProphetTable['weekday']*forecastProphetTable['yhat']
```

```python
forecastProphetTable.drop('weekday', axis=1, inplace= True)
```
